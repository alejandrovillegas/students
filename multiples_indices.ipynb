```
df = pd.melt(df_pob, id_vars=['Country Name'], var_name='year', value_name='pob')
df = df.sort_values(by=['year','Country Name'])
df
```

```
df['year'] = pd.Categorical(df['year'].apply(str))
df.dtypes
```

```
df_sample = df[idx_filtro]
df_sample
```
